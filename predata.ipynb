# Plots of sky-data related quantities that aren't yet being simultaneously calibrated

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import alan_data as adata
import notebook_utils as nutl

plt.ion()

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
nutl.setup_mpl()
```

```python
fig, ax = plt.subplots(2,1, sharex=True, figsize=(nutl.single_width*1.3, 0.75*nutl.single_width*1.3))

ax[0].plot(adata.sky_freq, adata.bmcorr, label=r'$\bar{b}_{\rm corr}$')
ax[0].plot(adata.sky_freq, adata.loss, label="L")
ax[0].plot(adata.sky_freq, adata.bmcorr*adata.loss, ls='--', label=r'L $\bar{b}_{\rm corr}$')
ax[1].plot(adata.sky_freq, adata.loss_temp)
ax[-1].set_xlabel("Frequency [MHz]")
ax[0].set_ylabel("Correction")
ax[0].legend(ncol=3,  labelspacing=0.25, handletextpad=0.5, columnspacing=1.5, handlelength=1.4, frameon=False)
ax[1].set_ylabel(r"$T_{\rm loss}$ [K]")
plt.tight_layout()
plt.savefig("plots/losses.pdf")
```
